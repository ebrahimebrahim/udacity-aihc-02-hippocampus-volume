## Loading NIFTI images using NiBabel

NiBabel is a python library for working with neuro-imaging formats (including NIFTI) that we have used in some of the exercises throughout the course. Our volumes and labels are in NIFTI format, so we will use nibabel to load and inspect them.

NiBabel documentation could be found here: https://nipy.org/nibabel/

Our dataset sits in two directories - *images* and *labels*. Each image is represented by a single file (we are fortunate to have our data converted to NIFTI) and has a corresponding label file which is named the same as the image file.

Note that our dataset is "dirty". There are a few images and labels that are not quite right. They should be quite obvious to notice, though. The dataset contains an equal amount of "correct" volumes and corresponding labels, and you don't need to alter values of any samples in order to get the clean dataset.

````python
# TASK: Your data sits in directory /data/TrainingSet.
# Load an image and a segmentation mask into variables called image and label

image = nib.load('../../data/TrainingSet/images/hippocampus_017.nii.gz')
label = nib.load('../../data/TrainingSet/labels/hippocampus_017.nii.gz')
````

````python
# Nibabel can present your image data as a Numpy array by calling the method get_fdata()
# The array will contain a multi-dimensional Numpy array with numerical values representing voxel intensities. 
# In our case, images and labels are 3-dimensional, so get_fdata will return a 3-dimensional array. You can verify this
# by accessing the .shape attribute. What are the dimensions of the input arrays?

image_array = image.get_fdata()
label_array = label.get_fdata()
print("Image shape is {} and dtype is {}, label shape is {} and dtype is {}".format(
    image_array.shape,image_array.dtype,label_array.shape,label_array.dtype
))
print("label values: {}".format(np.unique(label_array)))

# TASK: using matplotlib, visualize a few slices from the dataset, along with their labels. 
# You can adjust plot sizes like so if you find them too small:
plt.rcParams["figure.figsize"] = (6,6)

print("Some axial slices:")
slice_indices = [8,16,24]
fig, axs = plt.subplots(len(slice_indices),2)
for i,slice_index in enumerate(slice_indices):
    axs[i,0].imshow(image_array[:,:,slice_index],cmap='gray')
    axs[i,1].imshow(label_array[:,:,slice_index],cmap='gray')
    axs[i,0].axis('off')
    axs[i,1].axis('off')
    axs[i,0].set_title("z={}".format(slice_index))
````

````python
print("Some coronal slices:")
slice_indices = [8,16,24]
fig, axs = plt.subplots(len(slice_indices),2)
for i,slice_index in enumerate(slice_indices):
    axs[i,0].imshow(image_array[:,slice_index,:],cmap='gray')
    axs[i,1].imshow(label_array[:,slice_index,:],cmap='gray')
    axs[i,0].axis('off')
    axs[i,1].axis('off')
    axs[i,0].set_title("y={}".format(slice_index))
````

## Looking at single image data
In this section we will look closer at the NIFTI representation of our volumes. In order to measure the physical volume of hippocampi, we need to understand the relationship between the sizes of our voxels and the physical world.

````python
# Nibabel supports many imaging formats, NIFTI being just one of them. I told you that our images 
# are in NIFTI, but you should confirm if this is indeed the format that we are dealing with
# TASK: using .header_class attribute - what is the format of our images?

print(type(image.header))
print(type(label.header))
print('From the type of these headers we can see that nibabel indeed interepreted the files as NIFTI.')
````

Further down we will be inspecting .header attribute that provides access to NIFTI metadata. You can use this resource as a reference for various fields: https://brainder.org/2012/09/23/the-nifti-file-format/

````python
# TASK: How many bits per pixel are used?
print("The image is {} bits per pixel and the label is {} bits per pixel".format(
    image.header['bitpix'],
    label.header['bitpix']
))
````

````python
# TASK: What are the units of measurement?
print("Here is the xyzt_units byte: {:08b}".format(
    image.header['xyzt_units'].item()
))
print("This indicates that the spatial units of measurement are millimeters. See the markdown cell below for reasoning.")
````

The NIFTI header documentation is the code [here](https://nifti.nimh.nih.gov/pub/dist/src/niftilib/nifti1.h).

Here are the relevant excerpts:

(1)
```
   Bits 0..2 of xyzt_units specify the units of pixdim[1..3]
    (e.g., spatial units are values 1..7).
   Bits 3..5 of xyzt_units specify the units of pixdim[4]
    (e.g., temporal units are multiples of 8).
```

(2)
```
                               /*! NIFTI code for millimeters. */
#define NIFTI_UNITS_MM      2
```

````python
# TASK: Do we have a regular grid? What are grid spacings?
print("Here are pixdim[1], pixdim[2], and pixdim[3] from the NIFTI header:")
print(image.header['pixdim'][1:(3+1)])
print("Remember from the cell above that this is in mm.")
if len(np.unique(image.header['pixdim'][1:(3+1)]))==1:
    print("Yes this is a regular grid.")
else:
    print("It's not a regular grid.")
````

````python
# TASK: What dimensions represent axial, sagittal, and coronal slices? How do you know?
# ANSWER: See markdown cell below.
````

Fixing the first dimension represents taking a sagittal slice. Increasing index in the first dimension corresponds to moving rightward.

Fixing the second dimension represents taking a coronal slice. Increasing index in the second dimension corresponds to moving in the anterior direction (front, towards face).

Fixing the third dimension corresponds to taking an axial slice. Increasing index in the third dimension corresponds to moving in the superior direction (up, towards head).

I know this from my exploration of the data in 3D slicer.

````python
# By now you should have enough information to decide what are dimensions of a single voxel
# TASK: Compute the volume (in mm) of a hippocampus using one of the labels you've loaded. 
# You should get a number between ~2200 and ~4500


# Since each voxel is 1mm by 1mm by 1mm,
# computing the volume in mm^3 is just a matter of counting voxels in the marked region
vol_of_label = lambda label_array : (label_array!=0).sum()
vol_of_label(label_array)
````

## Plotting some charts

````python
import os
````

````python
# TASK: Plot a histogram of all volumes that we have in our dataset and see how 
# our dataset measures against a slice of a normal population represented by the chart below.

labels_path = '../../data/TrainingSet/labels'
filenames = os.listdir('../../data/TrainingSet/labels')

filenames_and_volumes = [(filename, vol_of_label(nib.load(os.path.join(labels_path,filename)).get_fdata()))
                         for filename in filenames]

volumes = np.array([pair[1] for pair in filenames_and_volumes])

plt.hist(volumes, bins=100)
plt.show()
````

````python
# Indeed there is an outlier. Let's find its filename so we can remove it:

[(f,v) for f,v in filenames_and_volumes if v==volumes.max()]
````

````python
# Let's have a look at this bad one

images_path = '../../data/TrainingSet/images'
bad_filename = [f for f,v in filenames_and_volumes if v==volumes.max()][0]

image = nib.load(os.path.join(images_path, bad_filename))
image_array = image.get_fdata()
label = nib.load(os.path.join(labels_path, bad_filename))
label_array = label.get_fdata()

print("Shapes:",image_array.shape,label_array.shape)
plt.imshow(image_array[:,:,16],cmap='gray')
plt.show()
print("Hmm how did that get there?!")
````

````python
# Let's look at the histogram again but without the outlier

volumes = np.delete(volumes,volumes.argmax())
plt.hist(volumes,bins = 8)
plt.title("Hippocampus volumes in training data set (mm^3)")
plt.show()
````

````python
# Just in case, let's have a look at *all* the labels in a middle sagittal slice...

# gets number in a filename
num_from_filename = lambda fn : fn.split('.')[0].split('_')[-1]

filenames.sort(key=lambda fn : int(num_from_filename(fn)))
fig, axs = plt.subplots(27,10,figsize=(15,50))
for filename,ax in zip(filenames,axs.flatten()):
    label_array = nib.load(os.path.join(labels_path,filename)).get_fdata()
    slice_index = label_array.shape[0]//2
    ax.imshow(label_array[slice_index,:,:],cmap='gray')
    ax.axis('off')
    ax.set_title(num_from_filename(filename))
````
